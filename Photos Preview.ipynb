```python
import numpy as np
from PIL import Image
import os
```

```python
# 使用numpy库中的load函数读取npz文件
data = np.load('/Users/wenzel/Downloads/Contests/Python/datasets/imagenet_val_25.npz', allow_pickle=True)
# 查看npz文件中键值名称
print(data.files)
```

```python
# 图像提取与保存
for i in range(len(data['X'])):
    # 使用PIL库中的Image.fromarray函数将读取到的图片数据转换为图片格式
    img = Image.fromarray(data['X'][i])
    # 使用os库中的join函数将转换后的图片保存到指定文件夹
    img.save(os.path.join('/Users/wenzel/Downloads/Contests/Python/datasets/images', str(i) + '.png'))
```

```python
from PIL import Image
import matplotlib.pyplot as plt


# 读取并显示所有图像
fig, axes = plt.subplots(5, 5, figsize=(15, 15))

for i in range(25):
    # 读取图像
    img = Image.open(f'datasets/images/{i}.png')

    # 计算子图的行和列索引
    row = i // 5
    col = i % 5

    # 在子图上显示图像
    axes[row, col].imshow(img)

    # 设置坐标轴刻度
    axes[row, col].set_xticks([0, 0.25 * img.width, 0.5 * img.width, 0.75 * img.width, img.width])
    axes[row, col].set_yticks([0, 0.25 * img.height, 0.5 * img.height, 0.75 * img.height, img.height])
    
    # 设置坐标轴标签
    axes[row, col].set_xticklabels([0, 0.25, 0.5, 0.75, 1])
    axes[row, col].set_yticklabels([0, 0.25, 0.5, 0.75, 1])

plt.show()
```
